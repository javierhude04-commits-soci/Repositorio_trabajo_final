# PROYECTOS DE LEY 2021 - 2026

```python
# LIBRERÍAS (Añadimos Select para manejar menús desplegables)
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# CONFIGURAR DRIVER
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--window-size=1920,1080')

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
# ABRIR PÁGINA
url = "https://wb2server.congreso.gob.pe/spley-portal/#/expediente/search"
driver.get(url)
driver.maximize_window()
wait = WebDriverWait(driver, 20)

# Esperar que la tabla cargue inicialmente
wait.until(EC.presence_of_element_located((By.TAG_NAME, "table")))

# --- NUEVA LÓGICA ---
data_total = []
continuar = True
pagina_actual = 1

print("Iniciando captura total de proyectos (2021-2026)...")

while continuar:
    print(f"Extrayendo datos de la página {pagina_actual}...")

    # Esperar y extraer las filas de la tabla actual
    wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "tbody tr")))
    filas = driver.find_elements(By.CSS_SELECTOR, "tbody tr")

    for fila in filas:
        columnas = fila.find_elements(By.TAG_NAME, "td")
        if len(columnas) >= 6:
            data_total.append({
                "proyecto_ley": columnas[0].text,
                "fecha": columnas[1].text,
                "titulo": columnas[2].text,
                # ... (resto de tus columnas)
            })

    # Lógica de "Siguiente" automática
    try:
        boton_next = driver.find_element(By.CSS_SELECTOR, "button.p-paginator-next")

        # Verificamos si el botón está deshabilitado
        if "p-disabled" in boton_next.get_attribute("class"):
            print("¡Hemos llegado a la última página!")
            continuar = False
        else:
            driver.execute_script("arguments[0].click();", boton_next)
            pagina_actual += 1
            time.sleep(2) # Pausa para no saturar el servidor del Congreso
    except:
        print("No se encontró el botón siguiente. Fin del proceso.")
        continuar = False

print(f"Proceso terminado. Total de proyectos recolectados: {len(data_total)}")
```

```python
# CREAR DATAFRAME Y LIMPIEZA
df = pd.DataFrame(data_total)
```

```python
import pandas as pd

# Creamos una copia para no alterar el original por error
df_limpio = df.copy()

# Limpiar 'proyecto_ley' y extraer el NÚMERO y el AÑO
#
df_limpio['proyecto_ley'] = df_limpio['proyecto_ley'].str.replace('Proyecto de Ley\n', '', case=False).str.strip()
df_limpio['año'] = df_limpio['proyecto_ley'].str.extract(r'/(\d{4})')

# Limpiar 'fecha'
#
df_limpio['fecha'] = df_limpio['fecha'].str.replace('Fecha de Presentación', '', case=False).str.strip()

# Limpiar 'titulo'
# Quitamos "Título\n" y el texto repetitivo "PROYECTO DE LEY QUE..."
df_limpio['titulo_limpio'] = df_limpio['titulo'].str.replace('Título\n', '', case=False).str.strip()
df_limpio['titulo_nlp'] = df_limpio['titulo_limpio'].str.replace('PROYECTO DE LEY (QUE|DE|PARA)\s', '', case=False, regex=True)
```

```python
# CLASIFICACIÓN
# Usamos expresiones regulares para capturar variaciones de las palabras
def clasificar_enfoque(texto):
    texto = str(texto).lower()
    # Diccionario de términos clave
    patrones_punitivos = r'penal|delit|pena|sancion|seguridad|polic|carcel|prision'
    patrones_rehab = r'reinser|rehabilit|penitenciar|inpe|resocializa|derechos humanos'

    import re
    if re.search(patrones_punitivos, texto):
        return "Punitivo"
    elif re.search(patrones_rehab, texto):
        return "Rehabilitador/Social"
    else:
        return "Otros/Administrativo"

df["enfoque"] = df["titulo"].apply(clasificar_enfoque)
```

```python
# RESULTADOS Y GUARDADO
print("\n--- RESUMEN DE HALLAZGOS ---")
print(f"Total proyectos scrapeados: {len(df)}")
print("\nDistribución por Enfoque:")
print(df["enfoque"].value_counts())

df.to_excel("proyectos_ley_2021_2026.xlsx", index=False)
driver.quit()
print("\nArchivo guardado exitosamente.")
```

```python
# Extraemos el año del formato
# El regex busca 4 números después de la barra '/'
df['año_ext'] = df['proyecto_ley'].str.extract(r'/(\d{4})')

print("--- DISTRIBUCIÓN CRONOLÓGICA (2021-2026) ---")
conteo_años = df['año_ext'].value_counts().sort_index()
print(conteo_años)

# Verificamos si hay algún 'NaN' (proyectos donde no se pudo leer el año)
nulos = df['año_ext'].isna().sum()
if nulos > 0:
    print(f"\nAtención: Hay {nulos} proyectos sin año identificado.")
```

# PROYECTOS DE LEY 2016-2021

```python
import requests
from bs4 import BeautifulSoup
import time # Importante para no saturar el servidor

base_url = "https://www2.congreso.gob.pe/Sicr/TraDocEstProc/CLProLey2016.nsf/Local%20Por%20Numero%20Inverso?OpenView"
headers = {'User-Agent': 'Mozilla/5.0'}

todos_los_proyectos = []

# Recorremos de 500 en 500 (1, 501, 1001, 1501...)
for inicio in range(1, 6001, 500):
    print(f"Extrayendo desde el registro {inicio}...")

    # Construimos la URL con el parámetro de inicio
    url_paginada = f"{base_url}&Start={inicio}"

    try:
        response = requests.get(url_paginada, headers=headers, timeout=10)
        soup = BeautifulSoup(response.text, 'html.parser')

        filas = soup.find_all('tr')
        conteo_en_pagina = 0

        for fila in filas:
            celda_titulo = fila.find('td', attrs={'width': '600'})
            if celda_titulo:
                titulo = celda_titulo.get_text(strip=True)
                enlace = fila.find('a')
                nro_proyecto = enlace.get_text(strip=True) if enlace else "Sin Número"

                # Filtramos filas vacías o encabezados
                if "/" in nro_proyecto:
                    todos_los_proyectos.append({
                        'numero': nro_proyecto,
                        'titulo': titulo
                    })
                    conteo_en_pagina += 1

        print(f"  -> Se sumaron {conteo_en_pagina} proyectos nuevos.")

        # Si la página trajo muy pocos resultados, es que llegamos al final
        if conteo_en_pagina == 0:
            break

        # Pausa de 1 segundo
        time.sleep(1)

    except Exception as e:
        print(f"Error en el registro {inicio}: {e}")
        break

print(f"\n--- PROCESO FINALIZADO ---")
print(f"Total de proyectos acumulados: {len(todos_los_proyectos)}")
```

```python
import pandas as pd

# LIMPIEZA Y FILTRADO DE SEGURIDAD
# Usamos un DataFrame para manejar mejor los datos de investigación
df = pd.DataFrame(todos_los_proyectos)

# Eliminamos posibles duplicados
df = df.drop_duplicates(subset=['numero'])

# CLASIFICACIÓN
def categorizar_proyecto(titulo):
    t = titulo.upper()
    # Variables
    rehab = ["REINSERCIÓN", "REHABILITACIÓN", "RESOCIALIZACIÓN", "PENITENCIARIO", "INPE" "TRABAJO", "INTERNO", "PRISIÓN"]
    punitivo = ["PENA", "DELITO", "SANCIÓN", "CRIMINAL", "CÓDIGO PENAL", "INPE" "SEGURIDAD CIUDADANA"]

    score_rehab = sum(1 for w in rehab if w in t)
    score_punit = sum(1 for w in punitivo if w in t)

    if score_rehab > score_punit: return "Enfoque Rehabilitador"
    if score_punit > score_rehab: return "Enfoque Punitivo"
    return "Otros / Administrativo"

df['categoria_investigacion'] = df['titulo'].apply(categorizar_proyecto)

# EXTRACCIÓN DEL AÑO
#
df['año'] = df['numero'].str.extract(r'/(\d{4})')

# FILTRADO FINAL (Aseguramos el periodo 2016-2021)
df = df[df['año'].isin(['2016', '2017', '2018', '2019', '2020', '2021'])]

# GUARDAR PARA GOOGLE COLAB / EXCEL
nombre_archivo = "base_datos_reinsercion_2016_2021.xlsx"
df.to_excel(nombre_archivo, index=False)

print(f"\n--- REPORTE FINAL ---")
print(f"Total proyectos limpios (2016-2021): {len(df)}")
print("\nDistribución por Enfoque:")
print(df['categoria_investigacion'].value_counts())
print("\nProyectos por Año:")
print(df['año'].value_counts().sort_index())
print(f"\nArchivo guardado como: {nombre_archivo}")
```
